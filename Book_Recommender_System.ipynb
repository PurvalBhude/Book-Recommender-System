## Popularity Based Recommender System

Condition:
- Books that has minimum 250 rating
- highest average rating
- Top 50 Books

```python
rating_with_books = ratings.merge(books,on="ISBN")
```

```python
rating_with_books
```

```python
rating_with_books["Book-Rating"].unique()
```

```python
rating_with_books.dtypes
```

```python
num_rating = rating_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating
```

```python
# Convert 'Book-Rating' column to numeric, handling non-numeric values
rating_with_books['Book-Rating'] = pd.to_numeric(rating_with_books['Book-Rating'], errors='coerce')

avg_rating = rating_with_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
popular = num_rating.merge(avg_rating,on="Book-Title")
popular
```

```python
popular = popular[popular['num_rating']>=250].sort_values('avg_rating',ascending=False)
popular.drop_duplicates(subset=['Book-Title'],inplace=True)
popular.shape
```

```python
popular
```

```python
popular = popular.merge(books,on="Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M',"Image-URL-S","Image-URL-L",'num_rating','avg_rating']]
```

```python
popular
```

## Recommender System

Conditons
- we will consider only users who have rated more than 150 books
- we will consider only books for which more than 25 people have rated
- we are using collabarative filtering algorithm

```python
x = rating_with_books.groupby('User-ID').count()['Book-Rating'] > 150

user_rated150 = x[x].index
user_rated150
```

```python
filter = rating_with_books[rating_with_books['User-ID'].isin(user_rated150)]
filter
```

```python
y = filter.groupby('Book-Title').count()['Book-Rating'] >= 25
famous = y[y].index
```

```python
final_rating = filter[filter['Book-Title'].isin(famous)]
final_rating.drop_duplicates()
final_rating
```

```python
pt = finale_rating = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
similarity.shape
similarity[0]
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))


        data.append(item)
    return data
```

```python
recommend('1st to Die: A Novel')
```
